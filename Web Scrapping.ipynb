```python
%%time
import time
import re
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from multiprocessing import Pool

class UzumScraper:
    def __init__(self):
        self.links = []

    def get_product_links(self, num_pages):
        for x in range(1, num_pages + 1):
            url_search = f'https://uzum.uz/uz/category/Mashhur--1?sorting=orders&ordering=descending&currentPage={x}'
            self.links.append(url_search)

    def scrape_product_data(self, url):
        try:
            options = webdriver.ChromeOptions()
            options.add_argument('--headless')  # Run the browser in the background
            driver = webdriver.Chrome(options=options)
            
            driver.get(url)
            time.sleep(15)  # Add a delay of 15 seconds for each page to load
            soup = BeautifulSoup(driver.page_source, 'lxml')
            driver.quit()  # Close the WebDriver after scraping to release resources

            # Extract product data
            link = url
            try:
                name = soup.find('h1', {'class': 'title'}).text.strip()
                category = soup.find_all('span', {'itemprop': 'name'})[2].text.strip()
                rate = soup.find('span', {'class': 'rating-value'}).text.strip()
                review_count = soup.find('span', {'class': 'reviews-info'}).text.strip()
                review_count = re.search(r'\d+', review_count.strip()).group()
                orders = soup.find('div', {'class': 'orders'}).text.strip()
                orders = re.search(r'\d+', orders.strip()).group()
                seller = soup.find('a', {'data-test-id': 'link__product-seller'}).text.strip()
                delivery_info = soup.find('span', {'class': 'slightly transparent regular'}).text.strip()
                delivery_time = delivery_info.split(',')[0]
                delivery_price = delivery_info.split(',')[1].strip()
                stock_status = soup.find('span', {'data-test-id': 'text__product-quantity'}).text.strip()
                price = soup.find('span', {'class': 'currency'}).text.strip().replace('\xa0', '')
                old_price = soup.find('span', {'class': 'currency old-price'}).text.strip().replace('\xa0', '')
                badge = soup.find('div', {'class': 'badge'}).text.strip()
                installments = soup.find('div', {'class': 'installment-text'}).text.strip()
                installments = re.sub(r'\D', '', installments)
                sales_in_this_week = soup.find('div', {'class': 'offer-text'}).text.strip()
                sales_in_this_week = re.findall(r'\d+', sales_in_this_week)[0]

            except AttributeError as e:
                # If any of the elements are not found, set the corresponding variable to an empty string
                name = ''
                category = ''
                rate = ''
                review_count = ''
                orders = ''
                seller = ''
                delivery_time = ''
                delivery_price = ''
                stock_status = ''
                price = ''
                old_price = ''
                badge = ''
                installments = ''
                sales_in_this_week = ''

            product_data = {
                'Name': name,
                'Category': category,
                'Rate': rate,
                'Review Count': review_count,
                'Orders': orders,
                'Seller': seller,
                'Delivery Time': delivery_time,
                'Delivery Price': delivery_price,
                'Stock Status': stock_status,
                'Price': price,
                'Old Price': old_price,
                'Badge': badge,
                'Installments': installments,
                'Sales in this Week': sales_in_this_week,
                'Link': link
            }

            return product_data

        except Exception as e:
            print(f"Error scraping {url}: {e}")
            return None

    def scrape_all_product_data(self):
        with Pool(processes=4) as pool:  # Adjust the number of processes as needed
            data = pool.map(self.scrape_product_data, self.links)
        
        data = [item for item in data if item is not None]  # Remove None values
        df = pd.DataFrame(data)
        return df


# Example usage:
uzum_scraper = UzumScraper()
uzum_scraper.get_product_links(num_pages=100)
df = uzum_scraper.scrape_all_product_data()
print(df)
```

## Data cleaning

We start from Delivery time

```python
data['Delivery Time'].value_counts()
```

```python
delivery_time = []

for time in data['Delivery Time']:
    delivery_time.append(time[0])
    
delivery_time[0:10]
```

```python
data['Delivery Time X'] = delivery_time
```

```python
data['Delivery Time X'].value_counts()
```

Delivery Price

```python
data['Delivery Price'].value_counts()
```

```python
delivery_price = []

for x in range(0,3793):
    x = 0
    delivery_price.append(x)
    
len(delivery_price)
```

```python
data['delivery_price'] = delivery_price
```

```python
data['stock'] = data['Stock Status'].apply(lambda x: ''.join(re.findall(r'\d+', x)))
```

```python
data.dtypes
```

```python
stock = []

for x in data['stock']:
    if x == '':
        x = 0
    elif x is str:
        int(x)
    stock.append(x)
```

```python
type(stock[0])
```

```python
data['stock'] = stock
```

```python
data['stock'] = data['stock'].astype(int)
```

```python
data.dtypes
```

```python
data['stock'].max()
```

```python
data['price'] = data['Price'].apply(lambda x: int(''.join(re.findall(r'\d+', x))) if re.findall(r'\d+', x) else None)
```

```python
data['price'].std()
```

```python
data['price'].isna().sum()
```

```python
data['old_price'] = data['Old Price'].apply(lambda x: int(''.join(re.findall(r'\d+', x))) if re.findall(r'\d+', x) else None)
```

```python
data.old_price
```

```python
data['Badge'].value_counts()
```

```python
data['price_diff'] = data['old_price']-data['price']
```
